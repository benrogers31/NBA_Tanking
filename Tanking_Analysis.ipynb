```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv').replace("CHH","NOP").replace("NJN","BRK")\
.replace("VAN","MEM").replace("WSB","WAS").replace("NOK","NOP").replace("CHA","CHO")\
.replace("NOH","NOP").replace("SEA","OKC")

df = df[['year','team','seed','made_finals']]
```

<h3> This is a method to check if a team has made the finals within a specified ammount of time</h3>

```python
def did_make_finals(year,team,length_of_time):
    
    curr_df = df[df["year"] > year][df["team"] == team][df["year"] <= length_of_time+year][df["made_finals"] == True]  

    if curr_df.empty:
        return False
    else:
        return True

    
did_make_finals(2017,"TOR",5)
```

<h3> Here we are going to create 5 states for our markov models for the data set </h3>
    

```python
list_of_states = []
for index, row in df.iterrows():
    #this is getting the number of games in that particular season
    if row['year'] < 1996:
        season_len = 26
    elif row['year'] < 2005:
        season_len = 28
    else:
        season_len = 29
    #getting the state
    state= int((season_len-row["seed"])/5)
    if row["made_finals"] == True:
        state = 6
        
    #storing the state
    list_of_states.append(state)

df["state"] = list_of_states
```

```python
list_of_states = df.state.unique()
list_of_teams = df.team.unique()
list_of_years = df.year.unique()
```

```python
transition_dict = dict.fromkeys(list_of_states)

for key in transition_dict.keys():
    transition_dict[key] = dict.fromkeys(list_of_states,0)

for team in list_of_teams:
    for year in list_of_years:
        
        if year == 2020:
            continue
        
        try:
            curr_state = int(df[df["year"] == year][df["team"] == team]["state"])
            next_state = int(df[df["year"] == year+1][df["team"] == team]["state"])
            
            print(f"0 {year} {team}: {curr_state} -> {next_state}")
            
            transition_dict[curr_state][next_state] += 1
            print(transition_dict)
        except:
            continue
       

        
        
    
```

```python
transition_probs = dict()

for key in transition_dict.keys():
    transition_probs[key] = dict.fromkeys(list_of_states,0)
    
for key in transition_dict.keys():
    factor=1.0/sum(transition_dict[key].values())
    for key2 in transition_dict[key].keys():
         transition_probs[key][key2] = factor*transition_dict[key][key2]
    
    
    
transition_probs
```

<h3> Now that we have the transition probabilities we will run simulations to see how long it takes for a team in a lower state to make a higher state </h3> 

```python
def get_next_state(curr_state): 
    probs_list = []
    for val in transition_probs[curr_state].values():
        probs_list.append(val)
    
    return int(np.random.choice(list_of_states, 1, replace=False, p=probs_list))
    


def simulate_x_years(start_state, years):
    state_list = []
    
    curr_state = start_state
    for year in range(0,years):
        next_state = get_next_state(curr_state)
        state_list.append(next_state)
        curr_state = next_state 
    
    return state_list



def probabilities_per_year(start_state, iterations, years, goal_state):
    
    years_achieved = [0] * years
    
    for i in range(1,iterations+1):
        curr_sim = simulate_x_years(start_state,years)
        
        for year in range(0,years):
            
            if curr_sim[year] >= goal_state:
                years_achieved[year] += 1
                
    probs = [x / iterations for x in years_achieved ]  
        
        
    return probs
probabilities_per_year(6,10000,10,6)     
```
